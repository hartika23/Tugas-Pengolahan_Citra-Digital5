```python
# 1. Import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 2. Load gambar
# Pastikan file "lily of the valley.jpg" sudah di-upload ke Colab
img = cv2.imread("lily of the valley.jpg", cv2.IMREAD_GRAYSCALE)

# Tampilkan gambar asli
plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.title("Gambar Asli (Grayscale)")
plt.axis("off")
plt.show()

# 3. Buat kernel low-pass filter (mean filter 3x3)
H = np.ones((3,3)) / 9
print("Kernel Low-pass (Mean Filter):")
print(H)

# 4. Lakukan konvolusi
Y = convolve2d(img, H, mode='same', boundary='fill', fillvalue=0)

# 5. Tampilkan hasil
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Input Gambar Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(Y, cmap='gray')
plt.title("Output Setelah Low-pass Filter")
plt.axis("off")

plt.show()
```

```python
# 1. Import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 2. Load gambar
# Pastikan file "lily of the valley.jpg" sudah di-upload ke Colab
img = cv2.imread("lily of the valley.jpg", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.title("Gambar Asli (Grayscale)")
plt.axis("off")
plt.show()

# 3. Buat kernel high-pass filter (contoh Laplacian 3x3)
H = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

print("Kernel High-pass (Edge Detection):")
print(H)

# 4. Lakukan konvolusi
Y = convolve2d(img, H, mode='same', boundary='fill', fillvalue=0)

# Normalisasi hasil agar bisa ditampilkan dengan baik
Y = np.clip(Y, 0, 255)

# 5. Tampilkan hasil
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Input Gambar Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(Y, cmap='gray')
plt.title("Output Setelah High-pass Filter")
plt.axis("off")

plt.show()
```

```python
# 1. Import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 2. Load gambar
# Pastikan file "lily of the valley.jpg" sudah di-upload ke Colab
img = cv2.imread("lily of the valley.jpg", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.title("Gambar Asli (Grayscale)")
plt.axis("off")
plt.show()

# 3. Definisikan kernel low-pass (mean filter 3x3)
low_pass = np.ones((3,3)) / 9

# 4. Definisikan kernel high-pass (Laplacian 3x3)
high_pass = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# 5. Buat kernel band-pass dengan kombinasi
# Band-pass = High-pass + Low-pass (menjaga frekuensi menengah)
band_pass = high_pass + low_pass

print("Kernel Band-pass:")
print(band_pass)

# 6. Lakukan konvolusi
Y = convolve2d(img, band_pass, mode='same', boundary='fill', fillvalue=0)

# Normalisasi hasil agar tetap dalam rentang intensitas
Y = np.clip(Y, 0, 255)

# 7. Tampilkan hasil
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Input Gambar Asli")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(band_pass, cmap='gray')
plt.title("Kernel Band-pass")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(Y, cmap='gray')
plt.title("Output Setelah Band-pass Filter")
plt.axis("off")

plt.show()
```
